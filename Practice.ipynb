# Subtask 3.1 Combine the Dataframe by Genres

```python
# So! Task is that we have to combine All genre of movies
```

```python
# Creating New Data Frame by Name df_by_genre

df_by_genre = df.loc[:,"CVotes10":"VotesnUS"]
df_by_genre[["genre_1","genre_2","genre_3"]] = df[["genre_1","genre_2","genre_3"]]
```

```python
df_by_genre
```

```python
# Now creating column as 'cnt|'
```

```python
df_by_genre["cnt"] = 1
```

```python
df_by_genre
```

```python
# Group the movies by individual genres
df_by_g1 =df_by_genre.groupby("genre_1").aggregate(np.sum)
df_by_g2 =df_by_genre.groupby("genre_2").aggregate(np.sum)
df_by_g3 =df_by_genre.groupby("genre_3").aggregate(np.sum)
```

```python
df_by_g1
```

```python
# Now let's add all g1 series in df_add
df_add = df_by_g1.add(df_by_g2,fill_value = 0)
df_add = df_add.add(df_by_g3,fill_value = 0)
df_add
```

```python
# Let's Extract the top 10 occurence

genre_top_10 = df_add.loc[df_add["cnt"]>10]
```

```python
genre_top_10
```

```python
# Take the mean for every column by dividing with cnt 
genre_top_10.iloc[:,0:-1]=genre_top_10.iloc[:,0:-1].divide(genre_top_10["cnt"],axis=0)
```

```python
genre_top_10
```

```python

genre_top_10.loc[:,"VotesM":"VotesnUS"]=round(genre_top_10.loc[:,"VotesM":"VotesnUS"],2)
```

```python

genre_top_10[genre_top_10.loc[:,"CVotes10":"CVotesnUS"].columns]=genre_top_10[genre_top_10.loc[:,"CVotes10":"CVotesnUS"].columns].astype(int)
```

```python
genre_top_10
```

# Subtask 3.2: Genre Counts!
 Now let's plot the bar graph ' genres vs cnt '

```python

sns.barplot( x ="cnt" , y = genre_top_10.index  , data = genre_top_10)
plt.show()
```
